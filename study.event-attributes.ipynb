````python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import json
````

````python
df = pd.read_csv('dataset.csv')
````

````python
event_df = df[['event_name', 'event_attributes']]
event_names = event_df['event_name'].unique()

for ev in event_names:
    print(f'===> Event: {ev}')
    event_df = df[['event_name', 'event_attributes']]
    event_df = event_df[event_df.event_name==ev]
    events_with_attributes = event_df.join(event_df['event_attributes'].apply(json.loads).apply(pd.Series))
    columns = list(events_with_attributes.columns)[2:]
    for col in columns:
        vals = list(events_with_attributes[col].unique())
        print(f"* Column: {col}")
        print(f"  Values: {vals}")
        print('==========')
````
